```python
# prompt: First convert '?' and rest of the values to integer

# Convert '?' to NaN
df['High School Average Mark\' numeric'] = df['High School Average Mark\' numeric'].replace('?', np.nan)

# Convert the rest to integers, coercing NaNs to NaNs
df['High School Average Mark\' numeric'] = pd.to_numeric(df['High School Average Mark\' numeric'], errors='coerce')
```

```python
#  For every marks as '?' or greater than 100 replace it by median


median_value = df[df['High School Average Mark\' numeric'] != '?']['High School Average Mark\' numeric'].median()
df['High School Average Mark\' numeric'] = df['High School Average Mark\' numeric'].replace('?', median_value)
df['High School Average Mark\' numeric'] = df['High School Average Mark\' numeric'].astype(float)
df.loc[df['High School Average Mark\' numeric'] > 100, 'High School Average Mark\' numeric'] = median_value
```

```python
# prompt: Finding the correlation between columns with First year persistence

# Calculate correlations with 'First year persistence'
correlations = df.corr()['FirstYearPersistence {1,0}']

# Print correlations
print(correlations)
```

```python
# Using Plotly express find the relation between First year gpa column for each value of First year persistence by grouping First year GPA into 5 groups as 0-1,1-2,2-3,3-4,4-5
# Convert 'First Term Gpa' numeric' to numeric, handling errors
df["First Term Gpa' numeric"] = pd.to_numeric(df["First Term Gpa' numeric"], errors='coerce')

# Create bins for First year GPA
bins = [0, 1, 2, 3, 4, 5]
labels = ['0-1', '1-2', '2-3', '3-4', '4-5']
df['GPA Group'] = pd.cut(df["First Term Gpa' numeric"], bins=bins, labels=labels, right=False)


# Create the Plotly Express figure
fig = px.histogram(df, x="GPA Group", color="FirstYearPersistence {1,0}",
                   barmode="group", title="First Year Persistence by GPA Group")

# Update layout for better presentation
fig.update_layout(
    xaxis_title="First Year GPA Group",
    yaxis_title="Count",
    legend_title="First Year Persistence"
)

fig.show()
```

```python
# the same way do it for second column

# Convert 'Second Term Gpa' numeric' to numeric, handling errors
df["Second Term Gpa' numeric"] = pd.to_numeric(df["Second Term Gpa' numeric"], errors='coerce')

# Create bins for Second year GPA
bins = [0, 1, 2, 3, 4, 5]
labels = ['0-1', '1-2', '2-3', '3-4', '4-5']
df['GPA Group'] = pd.cut(df["Second Term Gpa' numeric"], bins=bins, labels=labels, right=False)

# Create the Plotly Express figure
fig = px.histogram(df, x="GPA Group", color="FirstYearPersistence {1,0}",
                   barmode="group", title="First Year Persistence by Second Term GPA Group")

# Update layout for better presentation
fig.update_layout(
    xaxis_title="Second Term GPA Group",
    yaxis_title="Count",
    legend_title="First Year Persistence"
)

fig.show()
```

```python


# Assuming 'FirstYearPersistence {1,0}' is the target variable
target_variable = 'FirstYearPersistence {1,0}'

# Iterate over relevant columns and create plots
for column in df.columns:
  if column != target_variable:
    fig = px.histogram(df, x=column, color=target_variable,
                       barmode="group", title=f"Distribution of {column} by First Year Persistence")
    fig.show()


```

```python
#a violin plot between first second and last column

import plotly.graph_objects as go

# Assuming the first, second, and last columns are named as follows:
first_column = df.columns[0]
second_column = df.columns[1]
last_column = df.columns[-1]

fig = go.Figure()

# Add violin plot for the first column
fig.add_trace(go.Violin(y=df[first_column], name=first_column,
                        box_visible=True, meanline_visible=True))

# Add violin plot for the second column
fig.add_trace(go.Violin(y=df[second_column], name=second_column,
                        box_visible=True, meanline_visible=True))

# Add violin plot for the last column
fig.add_trace(go.Violin(y=df[last_column], name=last_column,
                        box_visible=True, meanline_visible=True))

# Update layout for better presentation
fig.update_layout(title="Violin Plot of First, Second, and Last Columns",
                  yaxis_title="Value")

fig.show()
```

```python
# prompt: Using -4 and -5 columns make a KDE plot with First year persistence

# Assuming -4 and -5 columns are named as follows:
column_minus_4 = df.columns[-4]
column_minus_5 = df.columns[-5]

fig = go.Figure()

# Add KDE plot for the -4 column
fig.add_trace(go.Violin(x=df['FirstYearPersistence {1,0}'], y=df[column_minus_4],
                        name=column_minus_4, box_visible=True, points="all",
                        meanline_visible=True))

# Add KDE plot for the -5 column
fig.add_trace(go.Violin(x=df['FirstYearPersistence {1,0}'], y=df[column_minus_5],
                        name=column_minus_5, box_visible=True, points="all",
                        meanline_visible=True))

# Update layout for better presentation
fig.update_layout(title="KDE Plot of -4 and -5 Columns by First Year Persistence",
                  xaxis_title="First Year Persistence",
                  yaxis_title="Value")

fig.show()
```
